```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
X = df[["cylinders", "displacement", "horsepower", "acceleration", "weight", "model_year"]]
y = df["mpg"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train_const = sm.add_constant(X_train)#adding constant to the x train value
ols_model = sm.OLS(y_train,X_train_const).fit()

```

```python
# Select significant features
significant_features = ols_model.pvalues[ols_model.pvalues < 0.5].index

# Remove 'const' if present
if 'const' in significant_features:
    significant_features = significant_features.drop('const')

# Refit model with significant features
X_train_sig_const = sm.add_constant(X_train[significant_features])
ols_model_sig = sm.OLS(y_train, X_train_sig_const).fit()

# Predict on test data
X_test_sig_const = sm.add_constant(X_test[significant_features])
y_pred_ols = ols_model_sig.predict(X_test_sig_const)
```

```python
y_pred_ols
```

```python
ols_model_sig.params
```

```python
y_pred_ols
```

```python
rmse_ols = np.sqrt(mean_squared_error(y_test, y_pred_ols))
```

```python
lr=LinearRegression()
lr.fit(X_train[significant_features],y_train)
y_pred_lr=lr.predict(X_test[significant_features])

rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
```

```python
print("significanrt fetaures",list(significant_features))
print("rmse values for ols method",rmse_ols)
print("rmse values for lr method",rmse_lr)
```

```python
final_df = X_test.copy()
final_df["mpg_value"]=y_test
final_df["mpg_value_lr"]=y_pred_lr
final_df["mpg_value_ols"]=y_pred_ols

```

```python
final_df.head()
```

```python
sns.scatterplot(final_df,x="mpg_value",y="mpg_value_ols",alpha=0.6,label="mpg vs ols_mpg _pred")
sns.scatterplot(final_df,x="mpg_value",y="mpg_value_lr",alpha=0.6,label="mpg vs lr_mpg _pred")

min_val = min(final_df['mpg_value'].min(), final_df['mpg_value_ols'].min(), final_df['mpg_value_lr'].min())
max_val = max(final_df['mpg_value'].max(), final_df['mpg_value_ols'].max(),final_df['mpg_value_lr'].max())
plt.plot([min_val, max_val], [min_val, max_val], 'k--', lw=2, label='Perfect Prediction')

# Labels and title
plt.xlabel('Actual MPG')
plt.ylabel('Predicted MPG')
plt.title('Actual vs Predicted MPG')
plt.legend()
plt.grid(True)
plt.show()
```
